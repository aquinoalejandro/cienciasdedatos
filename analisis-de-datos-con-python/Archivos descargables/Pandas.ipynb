## Analizando una base de datos

https://datos.bancomundial.org/indicator/SP.POP.TOTL

```python
dfpob=pd.read_csv('poblacion/API_SP.POP.TOTL_DS2_es_csv_v2_4149960.csv',skiprows=3)
display(dfpob.head())
```

```python
dfpob.set_index('Country Code',inplace=True)
```

```python
dfpob.head()
```

```python
dfpob2=dfpob[[str(k) for k in range(1960,2021)]]
display(dfpob2)
```

```python
paises=['MEX','PER','COL','ESP']
dfpob4=dfpob2.loc[paises]
display(dfpob4)
plt.figure(figsize=(12,6))
for pais in paises:
    plt.plot(dfpob4.loc[pais],label=pais)
plt.legend()
plt.grid()
xvalues=np.arange(1960,2021,5)
xvalues=[str(k) for k in xvalues]
plt.xticks(xvalues)
plt.show()
```

```python
plt.figure(figsize=(12,6))
ancho_barras=0.35
plt.grid()
indice_barras=np.arange(4)
plt.bar(indice_barras,dfpob4['1960'],width=ancho_barras,label='1960')
plt.bar(indice_barras+ancho_barras,dfpob4['2020'],width=ancho_barras,label='2020')
plt.xticks(indice_barras+ancho_barras,dfpob4.index)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.title('1960')
plt.pie(dfpob4['1960'],labels=dfpob4.index,shadow=True,autopct='%1.1f%%')
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.title('2020')
plt.pie(dfpob4['2020'],labels=dfpob4.index,shadow=True,autopct='%1.1f%%')
plt.show()
```

## Combinando DataFrames

```python
General=pd.DataFrame({'nombre':['Pepito','Maria','Ana','Bob'],
                      'edades':[16,12,17,18],
                      'escolaridad':['P','S','U','U']})
display(General)

Cursos=pd.DataFrame({'nombre':['Maria','Ana','Pepito'],
                     'cursos':['Math','Science','Data Science'],
                     'calificacion':[9,8,10],
                     'asistencias':[25,24,26]})
display(Cursos)
```

```python
pd.merge(General[['nombre','edades']],Cursos[['nombre','asistencias',]],how='outer')
```

```python
Estudiantes1=pd.DataFrame({'nombre':['Pepito','Maria','Ana'],
                      'edades':[16,12,17]})
display(Estudiantes1)
Estudiantes2=pd.DataFrame({'nombre':['Juan','Pedro','Bob'],
                      'edades':[15,14,17]})
display(Estudiantes2)
```

```python
pd.concat([Estudiantes1,Estudiantes2])
```

## Funciones y mapping

```python
Datos=pd.DataFrame({'Matematica':[10,8,7],
                    'Fisica':[6,5,9],
                    'Programacion':[7,9,10],
                   },index=['Pepito','Ana','Maria'])
display(Datos)

matematica=Datos['Matematica']
display(matematica)
```

```python
def restauno(n):
    return n-1

matematica.apply(restauno)
```

```python
Datos.applymap(restauno)
```

```python
def promedio(lista):
    return sum(lista)/len(lista)

Datos.apply(promedio,axis=1)
```

```python
Datos.agg([min,max,promedio],axis=1)
```

## Tablas Pivote

```python
Tabla=pd.DataFrame({
    'tipo':['moto','auto','moto','moto','auto','auto','moto','moto'],
    'velocidad':[12,23,34,26,43,21,28,30],
    'propietario':['Juan','Jorge','Maria','Ana','Juan','Maria','Bob','Camilo'],
    'nuevo':['si','no','no','si','no','si','si','no'],
    'precio':[5000,6000,7500,5500,8000,4000,6500,10000]
})
display(Tabla)
```

```python
Tabla.pivot_table(index='tipo',columns='nuevo',values='precio',aggfunc='max')
```

```python
Tabla.pivot_table(index='tipo',columns='nuevo',values='velocidad',aggfunc='min')
```

```python
Tabla.pivot_table(index='tipo',columns='nuevo',values='velocidad',aggfunc='count')
```

```python
Tabla.pivot_table(index='propietario',columns='tipo',values='velocidad',aggfunc='count').fillna(0)
```
